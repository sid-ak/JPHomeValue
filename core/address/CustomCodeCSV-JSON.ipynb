```python
import json
import csv 
import pandas as pd

df = pd.read_csv('/content/testdata.csv')


# Declaration of structure and initialization
data = {"addresses": [{"current_address":[{"index":[{"date":"01-01-2002", "price": "150000"}]}]}]}
# Outer loop - iterates over houses
for i in range(df.shape[1]-1):
  
  house_address = df.columns.values[3+i]
  #print(i)
  print(house_address)
  # Inner loop iterates over each data point for a specific house
  data["addresses"].append({house_address:[{0:[{"date":df['DATE'][0], "price":df[house_address][0]}]}]})
  for j in range(df.shape[0]-1):
    #print(j)
    #print('Populate First Price Entry')
    #data["addresses"].append({house_address:[{0:[{"date":df['DATE'][0], "price":df[house_address][0]}]}]})
    #print('Populating with the following price entries')
    data["addresses"][i+1][house_address].append({(j+1):[{"date":df['DATE'][j+1], "price":df[house_address][j+1]}]})
      
    with open("output10.json", "w") as f:
      json.dump(data, f, indent=4)
```
