```python
#imports
import firebase_admin
from firebase_admin import credentials
from firebase_admin import db
import csv
import json
```

```python
#function converting from csv to json
def convertAddressCsvToJson(csvFilePath, jsonFilePath):
  i = 0
  #step 1: create a directory
  data = {}

  #step 2: create csv reader
  with open(csvFilePath, encoding = 'utf-8') as csvf:
    csvReader = csv.DictReader(csvf)
    for rows in csvReader:
      key = i
      data[key] = rows
      i = i + 1

  #step 3: create a json writer
  with open(jsonFilePath, 'w', encoding = 'utf-8') as jsonf:
    jsonf.write(json.dumps(data, indent = 4))
```

```python
#connecting to database for writing
def connectToDb():
  #use your own credentials
  cred = firebase_admin.credentials.Certificate('Path to credentials json goes here')
  init_app = firebase_admin.initialize_app(cred, {
      'databaseURL': "https://jphomevalue-default-rtdb.firebaseio.com/" 
  })
```

```python
#function for writing to clearwater address data directory in db
def writeClearwaterData(jsonFilePath):
  #creating reference to clearwater address data
  ref = db.reference('/test/AddressData/ClearwaterAddressData/addressData') #/test is test branch on db / is main
  #writing to database
  with open(jsonFilePath, "r") as f:
    file_contents = json.load(f)
  ref.set(file_contents)

  print(file_contents)
```

```python
#function for writing to st pete address data directory in db
def writeStpData(jsonFilePath):
  #creating reference to St Pete address data
  ref = db.reference('/test/AddressData/StPeteAddressData/addressData') #/test is test branch on db / is main
  #writing to database
  with open(jsonFilePath, "r") as f:
    file_contents = json.load(f)
  ref.set(file_contents)

  print(file_contents)
```

```python
#function for writing to tampa address data directory in db
def writeTampaData(jsonFilePath):
  #creating reference to St Pete address data
  ref = db.reference('/test/AddressData/TampaAddressData/addressData') #/test is test branch / is main
  #writing to database
  with open(jsonFilePath, "r") as f:
    file_contents = json.load(f)
  ref.set(file_contents)

  print(file_contents)
```

```python
#driver code

#connect to db
if not firebase_admin.get_app():
  connectToDb()

#establish clearwater paths
csvFilePath = r'/home/jpmorgan/CWA.csv' #path to csv that you'd like to convert
jsonFilePath = r'/home/jpmorgan/CWA.json' #resulting json path

#create clearwater json
convertAddressCsvToJson(csvFilePath, jsonFilePath)

#write clearwater data to db
writeClearwaterData(jsonFilePath)

#establish st pete paths
csvFilePath = r'/home/jpmorgan/SPD.csv' #path to csv that you'd like to convert
jsonFilePath = r'/home/jpmorgan/SPD.json' #resulting json path

#create st pete json
convertAddressCsvToJson(csvFilePath, jsonFilePath)

#write st pete data to db
writeStpData(jsonFilePath)

#establish tampa paths
csvFilePath = r'/home/jpmorgan/TD.csv' #path to csv that you'd like to convert
jsonFilePath = r'/home/jpmorgan/TD.json' #resulting json path

#create tampa json
convertAddressCsvToJson(csvFilePath, jsonFilePath)

#write tampa data to db
writeTampaData(jsonFilePath)
```
